# Handwritten Digit Recognition

* Handwritten digits from MNIST datase
* 60,000 training set and 10,000 test sets
* Each example is a 28x28 grayscale image, associated with a label from 10 classes

Train a single layer neural network
1. 512-neurons hidden layer (ReLU)
2. output layer (softmax)

## Training with maximum of 10 ephochs and cutoff accuracy of 99 %

```python
import tensorflow as tf
mnist = tf.keras.datasets.mnist

# Set callbacks
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('acc')>0.99):
      print("\nReached 99% accuracy so cancelling training!")
      self.model.stop_training = True

callbacks = myCallback()

#obtain MNIST train and test sets
(x_train, y_train),(x_test, y_test) = mnist.load_data()
dimi = len(x_train[0])
dimj = len(x_train[0][0])

# Normalization of data
x_train = x_train / 255.0
x_test = x_test / 255.0

# NN construction
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(dimi, dimj)),
    tf.keras.layers.Dense(512, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# Optimizer and Loss function
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Training
model.fit(x_train, y_train, epochs=10, callbacks=[callbacks])
```

## Testing

```python
model.evaluate(x_test, y_test)
```
